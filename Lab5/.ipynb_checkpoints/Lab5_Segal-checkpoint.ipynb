```python
import pandas as pd
import numpy as np

df = pd.read_csv('acs2017_census_tract_data.csv', encoding='unicode_escape', low_memory=False) # read in the csv file

# removing all missing data
df.replace('', np.nan, inplace=True)
df.dropna(inplace=True)

#drop the county, ID sections because do not give us any contribution data
df = df.drop(['County','TractId'], axis=1)

df.sample(5)
```

```python
import collections

# make another column contains what we have more men/women

more_men = [] # contain 0 for more women , 1 for more men, 2 for pretty even
for m,w in zip(df.Men,df.Women):
    tot = m+ w
    if m > 0.53 * tot:
        more_men.append('m')
    elif m < 0.47 * tot:
        more_men.append('w')
    else:
        more_men.append('eq')
        
more_men = np.array(more_men)
collections.Counter(more_men)

    
```

```python
# make column for the majority race

major_race = [] # 0 more black, 1 more white, 2 more hispanic, 3 native, 4 asian
for w,b,h,n,a in zip(df.White,df.Black,df.Hispanic,df.Native,df.Asian):
    if max(w,b,h,n,a) == b:
        major_race.append('b')
    elif max(w,b,h,n,a) == w:
        major_race.append('w')
    elif max(w,b,h,n,a) == h:
        major_race.append('h')
    elif max(w,b,h,n,a) == n:
        major_race.append('n')
    else:
        major_race.append('a')
        

major_race = np.array(major_race)
print(collections.Counter(major_race))

        
```

```python
# make col for area of occupation column:

occupation = []

for prfs,s,o,c,prdc in zip(df.Professional,df.Service,df.Office,df.Construction,df.Production):
    if max(prfs,s,o,c,prdc) == prfs:
        occupation.append('prfs')
    elif max(prfs,s,o,c,prdc) == s:
        occupation.append('s')
    elif max(prfs,s,o,c,prdc) == o:
        occupation.append('o')
    elif max(prfs,s,o,c,prdc) == c:
        occupation.append('c')
    else:
        occupation.append('prdc')
        
occupation = np.array(occupation)
print(collections.Counter(occupation))
```

```python
# make categorical column for income

cat_income = []

per33 = np.percentile(df['Income'], 33)
per67 = np.percentile(df['Income'], 67)

for i in df.Income:
    if i < per33:
        cat_income.append('low')
    elif i < per67:
        cat_income.append('med')
    else:
        cat_income.append('high')
        
print('per33=',per33,'per67=',per67)
cat_income = np.array(cat_income)
print(collections.Counter(cat_income))
```

I am splitting the target of the child poverty into four categories:
* 0 - below 7%, not at risk at all
* 1 - below 18%, under the average child poverty rate in the U.S.[1], not at risk
* 2 - below 37%, above the avarage should be follewed 
* 3 - above 37%, third world conditions[2] - extreme poverty

Information where took from:

[1] https://en.wikipedia.org/wiki/Poverty_in_the_United_States#Child_poverty

[2] https://www.unicef.org/socialpolicy/files/child_poverty_in_the_developing_world.pdf

```python

y = pd.cut(df.ChildPoverty,bins=[-1,7,18,37,100], include_lowest=False,labels=[0,1,2,3]).to_numpy()
collections.Counter(y)
```

```python
# drop child poverty from the data
X = df.drop(['ChildPoverty','State'], axis=1)

# Normalizing the numeric data [-1,1]
X = -1 + ((X-X.min())*2/(X.max()-X.min())) 

# add the colums I created
X['states'] = df.State
X['more_men'] = more_men
X['major_race'] = major_race
X['occupation'] = occupation
X['cat_income'] = cat_income

#shape[1] should be 33 + 5 new categorical columns
print(type(X))
X.shape
```

The final data set I am using contains 33 numerical attributes plus 5 categorical columns I made for the states, gender, occupation, race, and income. I deleted the original columns of the ID and the county since they do not contribute any important data. Also, I deleted the missing data in all columns. For the target set, I categorized the data into four groups as shown above. Finally, I normalize the numeric values in the data set on a scale of -1 to 1. I did not one hot encoded the categorical data since Keras will do it for me later. After splitting the data into train and test sets, I will separate again the numeric values and the categorical columns to train the deep neural network.

### Cross product features

Using the catagorical colums I made, I will cross it into the following groups:

- State - race: because there relation between the state and the major race in it
- Occupation - gender: there are areas where is more common to one of the genders, or both
- Race - income: Statisticaly, there is a connection between higher average income and race[3] 
- Occupation - income: there are areas where the salaried are higher on average
- Occupation - race: There are areas where are more common to spesific race[4]



[3] https://en.wikipedia.org/wiki/List_of_ethnic_groups_in_the_United_States_by_household_income

[4] https://www.bls.gov/opub/reports/race-and-ethnicity/2017/home.htm



### Performance Evaluation

When I splinted the target into four different grouped, I compare those points according to the average child poverty in the U.S. as well the average in third world countries. The difference between the first and the second groups (0 and 1) was made just in order to split the data equally. Therefore, false positive or true negative between those two categories should not affect the evaluation of my algorithm. However, the difference between the second and the third group (1 and 2), from lower than the U.S average to a higher rate and the difference between the third and the last group (2 and 3), moving from “watch” group to the extreme poverty group will be a heavy weighted effect. With that being said, for those differences, predicting an area to be in a poorer group than it actually is, is less weighted than predicting an area to be in a “richer” group because it is better to be in a “safer mode” where you take more care even if it is not necessary, other than not taking care of an area that actually need it. 

In order to evaluate my performance I will use the weight table below, where the lower the vaules time the related weight the algorithm get, the better it is:



|      |Pred 0|Pred 1|Pred 2|Pred 3|
|------|------|------|------|------|
|real 0|   0  |   1  |   4  |  8   |
|real 1|   1  |   0  |   3  |  5   |
|real 2|   8  |   6  |   0  |  3   |
|real 3|   10 |   8  |   6  |  0   |

### Train/Test dividing

Since the data contains more than 72K different elements, defined by 34 attributes, it is large enough dataset to have variation in the train set, which will conatain more than 58K elements, as well as in the test one which will have more than 14K elements.  if we shuffle it and choose our 20% test set randomly, it will be most likely the same result as for any other general case. 

If we would want to be the most accurate with evaluation our model, we should use cross validation. However, since our dataset is large enough, the cost we will pay for computing the 'safer way' won't worth the little change it will make in our evaluation in the general case.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.20,
                                                    random_state=42,
                                                    stratify=y,
                                                    shuffle = True)

Y_train_ohe = pd.get_dummies(Y_train).values
Y_test_ohe = pd.get_dummies(Y_test).values


X_train.sample(10)
```

```python
collections.Counter(Y_test)
```

I split it so the test target will be with the same categorioes ratio as the whole dataset, using the stratify=y variable.

```python
from sklearn import metrics as mt
import keras

keras.__version__
```

```python
from keras.layers import Dense, Activation, Input
from keras.layers import Embedding, Flatten, Concatenate
from keras.models import Model
from keras.layers import concatenate
from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot
```

### Model 1:

for this model I will use all the 5 pairs I mentions before for the wide part of the network.
In the deep part, I will use 3 layer proceptron. the first will have 50 neurons, the second 30 and the lasst 10. Both part will be combained together to a layer with 4 neorons which is the dimension of the output y.

```python

from sklearn.preprocessing import LabelEncoder

encoders = dict() 
categorical_headers  = ['states','more_men','major_race','occupation','cat_income']
# numeric header are all the other
numeric_headers = list(set(X.columns.values) - set(categorical_headers))
print(numeric_headers)

for col in categorical_headers:
    
    # integer encoded variables
    encoders[col] = LabelEncoder() 
    X_train[col+'_int'] = encoders[col].fit_transform(X_train[col])
    X_test[col+'_int'] = encoders[col].transform(X_test[col])

```

```python
X_train_num =  X_train[numeric_headers].values
X_test_num = X_test[numeric_headers].values


# make the pairs
cross_columns = [['states','major_race'],
                ['occupation','more_men'],
                ['major_race','cat_income'],
                ['occupation','cat_income'],
                ['occupation','major_race']]


X_ints_train = [] 
X_ints_test = []
all_inputs = [] 
all_wide_branch_outputs = [] 

for cols in cross_columns:
    # encode crossed columns as ints for the embedding
    enc = LabelEncoder()
    
    # create crossed labels
    X_crossed_train = X_train[cols].apply(lambda x: '_'.join(x), axis=1)
    X_crossed_test = X_test[cols].apply(lambda x: '_'.join(x), axis=1)
    
    enc.fit(np.hstack((X_crossed_train.values,  X_crossed_test.values)))
    X_crossed_train = enc.transform(X_crossed_train)
    X_crossed_test = enc.transform(X_crossed_test)
    X_ints_train.append( X_crossed_train )
    X_ints_test.append( X_crossed_test )
    
    # get the number of categories
    N = max(X_ints_train[-1]+1) 
    
    # create embedding branch from the number of categories
    inputs = Input(shape=(1,),dtype='int32', name = '_'.join(cols))
    all_inputs.append(inputs)
    x = Embedding(input_dim=N, 
                  output_dim=int(np.sqrt(N)), 
                  input_length=1, name = '_'.join(cols)+'_embed')(inputs)
    x = Flatten()(x)
    all_wide_branch_outputs.append(x)
    
# merge the branches together
wide_branch = concatenate(all_wide_branch_outputs, name='wide_concat')
wide_branch = Dense(units=1,activation='sigmoid',name='wide_combined')(wide_branch)
```

```python
all_deep_branch_outputs = []

# add in the embeddings
for col in categorical_headers:
    # encode as ints for the embedding
    X_ints_train.append( X_train[col+'_int'].values )
    X_ints_test.append( X_test[col+'_int'].values )
    
    # get the number of categories
    N = max(X_ints_train[-1]+1) 
    
    # create embedding branch from the number of categories
    inputs = Input(shape=(1,),dtype='int32', name=col)
    all_inputs.append(inputs)
    x = Embedding(input_dim=N, 
                  output_dim=int(np.sqrt(N)), 
                  input_length=1, name=col+'_embed')(inputs)
    x = Flatten()(x)
    all_deep_branch_outputs.append(x)
    
# also get a dense branch of the numeric features
all_inputs.append(Input(shape=(X_train_num.shape[1],),
                        sparse=False,
                        name='numeric_data'))

x = Dense(units=20, activation='relu',name='numeric_1')(all_inputs[-1])
all_deep_branch_outputs.append( x )

# merge the deep branches together
deep_branch = concatenate(all_deep_branch_outputs,name='concat_embeds')
deep_branch = Dense(units=50,activation='relu', name='deep1')(deep_branch)
deep_branch = Dense(units=30,activation='relu', name='deep2')(deep_branch)
deep_branch = Dense(units=10,activation='relu', name='deep3')(deep_branch)
    
final_branch = concatenate([wide_branch, deep_branch],name='concat_deep_wide')
final_branch = Dense(units=4,activation='sigmoid',name='combined')(final_branch)

model1 = Model(inputs=all_inputs, outputs=final_branch)

model1.summary()
```

```python

SVG(model_to_dot(model1, dpi=50).create(prog='dot', format='svg'))
```

```python
model1.compile(optimizer='adagrad',
              loss='mean_squared_error',
              metrics=['accuracy'])

history1 = model1.fit(X_ints_train+ [X_train_num],
                    Y_train_ohe, 
                    epochs=10, 
                    batch_size=32, 
                    verbose=1,  # give feedback
                    validation_data = (X_ints_test + [X_test_num], Y_test_ohe))
```

```python
# function to make one hot encoded to regular categorical data
def ohe_to_non(ohe):
    returnList = []
    for raw in ohe:
        i = np.argmax(raw)
        returnList.append(i)
    return np.array(returnList)
        
```

```python
yhat = np.round(model1.predict(X_ints_test + [X_test_num]))
yhat_ = ohe_to_non(yhat)

confused1 = mt.confusion_matrix(Y_test,yhat_)
acc1 = mt.accuracy_score(Y_test,yhat_)

print(confused1,"\nacc:",acc1)
```

### Model 2:

For this model, I will use the same side part as for model one, but on the deep part I will add more layeres with more neurons in each. I will use 5 layers which will start from 100 neurons in the first layer and then will go down to 10 at the last layer of the deep part. 

```python

# merge the deep branches together for the second model using the wide part from the first model
deep_branch2 = concatenate(all_deep_branch_outputs,name='concat_embeds')
deep_branch2 = Dense(units=100,activation='relu', name='deep1')(deep_branch2)
deep_branch2 = Dense(units=70,activation='relu', name='deep2')(deep_branch2)
deep_branch2 = Dense(units=50,activation='relu', name='deep3')(deep_branch2)
deep_branch2 = Dense(units=30,activation='relu', name='deep4')(deep_branch2)
deep_branch2 = Dense(units=10,activation='relu', name='deep5')(deep_branch2)
    
final_branch2 = concatenate([wide_branch, deep_branch2],name='concat_deep_wide')
final_branch2 = Dense(units=4,activation='sigmoid',name='combined')(final_branch2)

model2 = Model(inputs=all_inputs, outputs=final_branch2)

model2.summary()
```

```python
model2.compile(optimizer='adagrad',
              loss='mean_squared_error',
              metrics=['accuracy'])

history2 = model2.fit(X_ints_train+ [X_train_num],
                    Y_train_ohe, 
                    epochs=10, 
                    batch_size=32, 
                    verbose=1,  # give feedback
                    validation_data = (X_ints_test + [X_test_num], Y_test_ohe))
```

```python
yhat = np.round(model2.predict(X_ints_test + [X_test_num]))
yhat_ = ohe_to_non(yhat)

confused2 = mt.confusion_matrix(Y_test,yhat_)
acc2 = mt.accuracy_score(Y_test,yhat_)

print(confused2,"\nacc:",acc2)
```

### Model 3

the deep part of this model will be the same as for model 1. For the wide part, I will make different crossed groups:
* states - race - occupation : since state relate to the race and the race relate to the occupation
* Occupation - gender - income : since occupaption relate to the gender and the income
* states - income : we did not have this group before but it make sense than the income in certain states higher then others

```python
# make the the crossed groups
cross_columns = [['states','major_race','occupation'],
                ['occupation','more_men','cat_income'],
                ['states','cat_income']]

# use the all_inputs list as before but change the crossed groups, which are the 5 first elements
# so I will make new all_inputs list start from the 6th elements, the same for X_int_train and test:

all_inputs3 = all_inputs[5:]
X_ints_train3 = X_ints_train[5:] 
X_ints_test3 = X_ints_test[5:]

#in the for loop I will add the crossed groups at the begining of each list.


# follow the same process as before for the wide part:

all_wide_branch_outputs = [] 

for cols in cross_columns:
    # encode crossed columns as ints for the embedding
    enc = LabelEncoder()
    
    # create crossed labels
    X_crossed_train = X_train[cols].apply(lambda x: '_'.join(x), axis=1)
    X_crossed_test = X_test[cols].apply(lambda x: '_'.join(x), axis=1)
    
    enc.fit(np.hstack((X_crossed_train.values,  X_crossed_test.values)))
    X_crossed_train = enc.transform(X_crossed_train)
    X_crossed_test = enc.transform(X_crossed_test)
    X_ints_train3.insert(0, X_crossed_train )
    X_ints_test3.insert(0, X_crossed_test )
    
    # get the number of categories
    N = max(X_ints_train3[0]+1) 
    
    # create embedding branch from the number of categories
    inputs = Input(shape=(1,),dtype='int32', name = '_'.join(cols))
    all_inputs3.insert(0,inputs) # insert in the begining.
    x = Embedding(input_dim=N, 
                  output_dim=int(np.sqrt(N)), 
                  input_length=1, name = '_'.join(cols)+'_embed')(inputs)
    x = Flatten()(x)
    all_wide_branch_outputs.append(x)
    
# merge the branches together
wide_branch3 = concatenate(all_wide_branch_outputs, name='wide_concat')
wide_branch3 = Dense(units=1,activation='sigmoid',name='wide_combined')(wide_branch3)
```

```python
final_branch3 = concatenate([wide_branch3, deep_branch],name='concat_deep_wide')
final_branch3 = Dense(units=4,activation='sigmoid',name='combined')(final_branch3)

model3 = Model(inputs=all_inputs3, outputs=final_branch3)
```

```python
model3.compile(optimizer='adagrad',
              loss='mean_squared_error',
              metrics=['accuracy'])

history3 = model3.fit(X_ints_train3+ [X_train_num],
                    Y_train_ohe, 
                    epochs=10, 
                    batch_size=32, 
                    verbose=1,  # give feedback
                    validation_data = (X_ints_test3 + [X_test_num], Y_test_ohe))
```

```python
yhat = np.round(model3.predict(X_ints_test3 + [X_test_num]))
yhat_ = ohe_to_non(yhat)

confused3 = mt.confusion_matrix(Y_test,yhat_)
acc3 = mt.accuracy_score(Y_test,yhat_)

print(confused3,"\nacc:",acc3)
```

## Comparison between the models:

First, I will defnie a function for calculating the weight of error for each model based on the evaluation metric I shoed above:

```python
def weigh_error(conf_mt):
    er_weight = [[0,1,4,8],
                 [1,0,3,5],
                 [8,6,0,3],
                 [10,8,6,0]]
    er_weight = np.array(er_weight)
    calc_weight = conf_mt * er_weight
    return np.sum(calc_weight)
```

```python
print('Model 1:' , weigh_error(confused1))
print('Model 2:' , weigh_error(confused2))
print('Model 3:' , weigh_error(confused3))
```

Like I mentioned before, the lower the sum of the vaules time the related weight, the better it is. We can see that all the models are pretty close to one another by their accuracy. With that being said, considering the the result above, we see that model 2 with five crossed pairs in the wide side and five layers network on the deep said, performed the best amoung the three.

### Comparing to a standard MLP

For this task, I will take the second model I built and compare it to a 5 layers MLP. The data I will use is the numeric columns and only the states column from the categorical columns, since the others I made based on the numeric data. Since it is now categorical, I will one hot encode it add horizontali stack it with the x_num train set and test set.

```python

# one hot encode
states_ohe_train = pd.get_dummies(X_train['states']).values
states_ohe_test = pd.get_dummies(X_test['states']).values

# combine
MLP_X_train = np.hstack((X_train_num,states_ohe_train))
MLP_X_test = np.hstack((X_test_num,states_ohe_test))

#build the model

inputs = Input(shape=(MLP_X_train.shape[1],))

x = Dense(units=100, activation='relu')(inputs)
x = Dense(units=60, activation='relu')(x)
x = Dense(units=30, activation='relu')(x)
x = Dense(units=10, activation='relu')(x)
predictions = Dense(4,activation='softmax')(x)

MLP_model = Model(inputs=inputs, outputs=predictions)
```

```python


MLP_model.compile(optimizer='adam',
              loss='mean_squared_error',
              metrics=['accuracy'])

history_MLP = MLP_model.fit(MLP_X_train,
                    Y_train_ohe, 
                    epochs=10, 
                    batch_size=32, 
                    verbose=1,  # give feedback
                    validation_data = (MLP_X_test, Y_test_ohe))
```

```python
yhat = np.round(MLP_model.predict(MLP_X_test))
yhat_ = ohe_to_non(yhat)

confused_MLP = mt.confusion_matrix(Y_test,yhat_)
acc_MLP = mt.accuracy_score(Y_test,yhat_)

print(confused_MLP,"\nacc:",acc_MLP)
```

```python
print('Model 2:' , weigh_error(confused2),'acc:',acc2)
print('MLP Model:' , weigh_error(confused_MLP),'acc:',acc_MLP)
```

We can see that according to my evaluation metric, the MLP model is performe a little bit worse than model 2. Also, if we will look at their accuracies we will see that model 2 got 1.6% higher accuracy than the MLP model.
